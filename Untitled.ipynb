```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("rotation_test_values.csv")

# Parse timestamps (mixed ISO formats)
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="mixed")

# Filter from timestamp onward
t0 = pd.to_datetime("2026-02-11 14:26:44.065517")
df = df[df["Timestamp"] >= t0].copy()

# Compute magnitude
df["mag"] = np.sqrt(df["Val1"]**2 + df["Val2"]**2 + df["Val3"]**2)
df = df[df["mag"] <= 0.5]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df["Timestamp"], df["mag"])
plt.xlabel("Time")
plt.ylabel("sqrt(Val1² + Val2² + Val3²)")
plt.title("Vector Magnitude vs Time")
plt.show()
```
